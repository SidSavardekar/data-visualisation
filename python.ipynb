```python
import pandas as pd
import numpy as np
```

```python
disease= pd.read_csv("Datafile_070.csv")
```

```python
disease.head()
```

```python
disease.info()
```

```python
disease.isnull().sum()
```

```python
disease["risk_score"] = disease["u"] * (1 + disease["CoV"])
```

```python
disease.describe()
```

```python
bins = [0, 1.0, 2.0, disease["risk_score"].max()]
labels = ["Low", "Medium", "High"]
disease["risk_category"] = pd.cut(disease["risk_score"], bins=bins, labels=labels, include_lowest=True)
```

```python
print(disease["risk_category"].value_counts())
```

```python
bins = [0, 0.3, 1, 5, disease["u"].max()]
labels = ["Negligible", "Low", "Medium", "High"]
disease["u_category"] = pd.cut(disease["u"], bins=bins, labels=labels, include_lowest=True)
```

```python
print(disease["u_category"].value_counts())
```

```python
disease.to_csv("D:/PowerBI_disease_data.csv", index=False)
```
